```python
import numpy as np
import matplotlib.pyplot as plt
```

## Overfit illustration

```python
testacc = np.loadtxt("testacc.txt")
testloss=np.loadtxt("testloss.txt")
```

```python
trainacc=np.loadtxt("trainaccuracy.txt")
trainloss=np.loadtxt("trainloss.txt")
```

```python
methods = ["LBFGSfullbatch","SGD","SDLBFGS"]
bs_grid = [0,64,64]
hs_grid = [20,0,0]
colors=['r','b','g']
```

## Over test set

```python
for i in range(3):
    bs = bs_grid[i]
    color_=colors[i]
    hs = hs_grid[i]
    meth = methods[i]    
    plt.figure(1, figsize = (12,12))
    if i ==1:
        lr = 0.001
        plt.plot(testacc[i],label='{}, lr = {}, bs = {}'.format(meth,lr,bs),color=color_)
        plt.plot(testloss[i],color=color_)
        
    elif bs != 0 and hs !=0:
        plt.plot(testacc[i],label='{}, bs = {}, hs = {}'.format(meth,bs,hs),color=color_)
        plt.plot(testloss[i],color=color_)
    elif bs != 0:
        plt.plot(testacc[i],label='{}, bs = {}'.format(meth,bs),color=color_)
        plt.plot(testloss[i],color=color_)
    elif hs != 0:
        plt.plot(testacc[i],label='{}, hs = {}'.format(meth,hs),color=color_)
        plt.plot(testloss[i],color=color_)
    else:
        plt.plot(testacc[i],label='{}'.format(meth),color=color_)
        plt.plot(testloss[i],color=color_)
plt.xlabel('Epochs',fontsize=12)
plt.ylabel('Test Loss                                                                          Test Accuracy',fontsize=13)
plt.title('Test loss and accuracy in FCNN, overfit phenomenon',fontsize = 20)
plt.legend(fontsize = 13)
plt.savefig("images/overfit_test_fcnn.jpg")
plt.figure (2)
```
