```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
testacc = np.zeros((4,10))
testacc[1:5,:] = np.loadtxt("LBFGS_new_minibatch_cnn_1_testacc.txt")
testacc[0,:] =np.loadtxt("LBFGS_new_minibatch_cnn_512_testacc.txt")

trainacc = np.zeros((4,10))
trainacc[1:5,:] = np.loadtxt("LBFGS_new_minibatch_cnn_1_trainacc.txt")
trainacc[0,:] =np.loadtxt("LBFGS_new_minibatch_cnn_512_trainacc.txt")


testloss = np.zeros((4,10))
testloss[1:5,:] = np.loadtxt("LBFGS_new_minibatch_cnn_1_testloss.txt")
testloss[0,:] =np.loadtxt("LBFGS_new_minibatch_cnn_512_testloss.txt")


trainloss = np.zeros((4,10))
trainloss[1:5,:] = np.loadtxt("LBFGS_new_minibatch_cnn_1_trainloss.txt")
trainloss[0,:] =np.loadtxt("LBFGS_new_minibatch_cnn_512_trainloss.txt")



```

```python
testacc
```

```python
bs_grid = [512,1024, 2048, 4098]
hs = 5
for i in range(0,4):
    bs = bs_grid[i]
    
    plt.figure(1, figsize = (12,10))
    plt.plot(testloss[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(2, figsize = (12,10))
    plt.plot(trainloss[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(3, figsize = (12,10))
    plt.plot(testacc[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(4, figsize = (12,10))
    plt.plot(trainacc[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(1)
plt.xlabel('Epochs')
plt.ylabel('Test Loss')
plt.title('Minibatch LBFGS Test loss in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("results_plot/mini_LBFGS_test_loss_fcnn.jpg")
plt.figure (2)
plt.xlabel('Epochs')
plt.ylabel('Train Loss')
plt.title('Minibatch LBFGS Train loss in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("results_plot/mini_LBFGS_train_loss_fcnn.jpg")
plt.figure (3)
plt.xlabel('Epochs')
plt.ylabel('Test Accuracy')
plt.title('Minibatch LBFGS Test accuracy in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("results_plot/mini_LBFGS_test_acc_fcnn.jpg")
plt.figure (4)
plt.xlabel('Epochs')
plt.ylabel('Train Accuracy')
plt.title('Minibatch LBFGS Train accuracy in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("results_plot/mini_LBFGS_train_acc_fcnn.jpg")
```
