```python
results = {}

results["Custom Model"] = []
```

### Baseline pytorch MSE

```python
def model_creation():
    model_1 = nn.Sequential(
    nn.Linear(2, 25),nn.ReLU(),
    nn.Linear(25, 25),nn.ReLU(),
    nn.Linear(25, 25),nn.ReLU(),
    nn.Linear(25, 25),nn.ReLU(),
    nn.Tanh(),nn.Linear(25,2)
    );
    return model_1
```

```python
epochs = 250
eta, batch_size = 0.1 , 100
validation_ratio , validation_rounds = 0.1 , 10
one_hot_labels= True
criterion = nn.MSELoss()
results["Custom Model"].append(evaluate_model_pytorch(model_creation, criterion,eta,nb_samples, epochs, batch_size, validation_ratio, validation_rounds,one_hot_labels,False))
```

### Mse implementation

```python
criterion = MSE()
optimizer = SGD(0.05)

epochs = 250
mini_batch_size = 100
validation_ratio = 0.1
validation_rounds = 10
one_hot_labels = True

results["Custom Model"].append(evaluate_model(Neural_network, hidden_layers, nb_samples, criterion, optimizer, epochs, mini_batch_size, validation_ratio, validation_rounds,one_hot_labels, False))
```

### Baseline pytorch cross entropy

```python
epochs = 250
eta, batch_size = 0.05 , 100
validation_ratio , validation_rounds = 0.1 , 10
one_hot_labels= False
criterion = nn.CrossEntropyLoss()
results["Custom Model"].append(evaluate_model_pytorch(model_creation, criterion,eta,nb_samples, epochs, batch_size, validation_ratio, validation_rounds,one_hot_labels,False))
```

###  cross entropy implementation 

```python
criterion = Crossentropy()
optimizer = SGD(0.05)

epochs = 250
mini_batch_size = 100
validation_ratio = 0.1
validation_rounds = 10
one_hot_labels = False

results["Custom Model"].append(evaluate_model(Neural_network, hidden_layers, nb_samples, criterion, optimizer, epochs, mini_batch_size, validation_ratio, validation_rounds,one_hot_labels,False))
```

```python
import pandas as pd
colnames = ['Train error', 'Train stdev',  'Validation error', 'Validation stdev', 'Test error','Test stdev']
pdResult = {}
for model_name in results:
    pdResult[model_name] = pd.DataFrame(results[model_name])
    pdResult[model_name].columns = colnames
```
