```python
import numpy as np
import matplotlib.pyplot as plt

```

# SMALL batch sizes

```python
testacc = np.loadtxt("LBFGS_new_minibatch_fcNN_testacc_2.txt")
trainacc = np.loadtxt("LBFGS_new_minibatch_fcNN_trainacc_2.txt")
testloss = np.loadtxt("LBFGS_new_minibatch_fcNN_testloss_2.txt")
trainloss = np.loadtxt("LBFGS_new_minibatch_fcNN_trainloss_2.txt")
```

```python
np.shape(testacc)
```

```python
grid = [0,3,7,8]
for i in (grid):
    if  i==0 :
        bs = 32
        hs = 3
    elif i ==3 :
        bs = 64
        hs = 3
        
    elif i==7 or i ==8:
        bs = 128
        hs = 6
    if i == 8:
        hs = 10
        
    plt.figure(1, figsize = (12,10))
    plt.plot(testloss[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(2, figsize = (12,10))
    plt.plot(trainloss[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(3, figsize = (12,10))
    plt.plot(testacc[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(4, figsize = (12,10))
    plt.plot(trainacc[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(1)
plt.xlabel('Epochs')
plt.ylabel('Test Loss')
plt.title('Minibatch LBFGS Test loss in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/mini_LBFGSNew_test_loss_fcnn.jpg")
plt.figure (2)
plt.xlabel('Epochs')
plt.ylabel('Train Loss')
plt.title('Minibatch LBFGS Train loss in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/mini_LBFGSNew_train_loss_fcnn.jpg")
plt.figure (3)
plt.xlabel('Epochs')
plt.ylabel('Test Accuracy')
plt.title('Minibatch LBFGS Test accuracy in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/mini_LBFGSNew_test_acc_fcnn.jpg")
plt.figure (4)
plt.xlabel('Epochs')
plt.ylabel('Train Accuracy')
plt.title('Minibatch LBFGS Train accuracy in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/mini_LBFGSNew_train_acc_fcnn.jpg")
```

# BIG batch sizes

```python

testacc2 = np.loadtxt("LBFGS_new_minibatch_fcNN_testacc_1.txt")
trainacc2 = np.loadtxt("LBFGS_new_minibatch_fcNN_trainacc_1.txt")
testloss2 = np.loadtxt("LBFGS_new_minibatch_fcNN_testloss_1.txt")
trainloss2 = np.loadtxt("LBFGS_new_minibatch_fcNN_trainloss_1.txt")
```

```python
grid = [0,2,3]
for i in (grid):
    if  i==0 :
        bs = 512
        hs = 15
    elif i ==2 :
        bs = 1024
        hs = 15
    else:
        bs = 1024
        hs = 25

    plt.figure(1, figsize = (12,10))
    plt.plot(testloss2[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(2, figsize = (12,10))
    plt.plot(trainloss2[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(3, figsize = (12,10))
    plt.plot(testacc2[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(4, figsize = (12,10))
    plt.plot(trainacc2[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(1)
plt.xlabel('Epochs')
plt.ylabel('Test Loss')
plt.title('Minibatch LBFGS Test loss in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/miniBIG_LBFGSNew_test_loss_fcnn.jpg")
plt.figure (2)
plt.xlabel('Epochs')
plt.ylabel('Train Loss')
plt.title('Minibatch LBFGS Train loss in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/miniBIG_LBFGSNew_train_loss_fcnn.jpg")
plt.figure (3)
plt.xlabel('Epochs')
plt.ylabel('Test Accuracy')
plt.title('Minibatch LBFGS Test accuracy in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/miniBIG_LBFGSNew_test_acc_fcnn.jpg")
plt.figure (4)
plt.xlabel('Epochs')
plt.ylabel('Train Accuracy')
plt.title('Minibatch LBFGS Train accuracy in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/miniBIG_LBFGSNew_train_acc_fcnn.jpg")
```
