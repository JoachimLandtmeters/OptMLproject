```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
testacc = np.loadtxt("LBFGS_new_minibatch_fcnn_1_testacc.txt")
trainacc = np.loadtxt("LBFGS_new_minibatch_fcnn_1_trainacc.txt")
testloss = np.loadtxt("LBFGS_new_minibatch_fcnn_1_testloss.txt")
trainloss = np.loadtxt("LBFGS_new_minibatch_fcnn_1_trainloss.txt")
```

```python
np.shape(testacc)
```

```python
grid = [0]
bs_grid = [32,128]
hs_grid=[3,6]

for i in (grid):
    bs = bs_grid[i]
    hs = hs_grid[i]
    plt.figure(1, figsize = (12,10))
    plt.plot(testloss[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(2, figsize = (12,10))
    plt.plot(trainloss[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(3, figsize = (12,10))
    plt.plot(testacc[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(4, figsize = (12,10))
    plt.plot(trainacc[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(1)
plt.xlabel('Epochs')
plt.ylabel('Test Loss')
plt.title('Minibatch LBFGS Test loss in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/mini_LBFGSNew_test_loss_fcnn.jpg")
plt.figure (2)
plt.xlabel('Epochs')
plt.ylabel('Train Loss')
plt.title('Minibatch LBFGS Train loss in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/mini_LBFGSNew_train_loss_fcnn.jpg")
plt.figure (3)
plt.xlabel('Epochs')
plt.ylabel('Test Accuracy')
plt.title('Minibatch LBFGS Test accuracy in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/mini_LBFGSNew_test_acc_fcnn.jpg")
plt.figure (4)
plt.xlabel('Epochs')
plt.ylabel('Train Accuracy')
plt.title('Minibatch LBFGS Train accuracy in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/mini_LBFGSNew_train_acc_fcnn.jpg")
```
