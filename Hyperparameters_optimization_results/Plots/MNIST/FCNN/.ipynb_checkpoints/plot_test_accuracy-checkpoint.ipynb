```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
testacc = np.loadtxt("testacc.txt")

```

```python
np.shape(testacc)
```

```python
methods = ["SGD","LBFGSfullbatch","LBFGSfullbatch", "LBFGSminibatch", "LBFGSminibatch", "LBFGSNewminibatch","SDLBFGS", "Curveball"]
bs_grid = [64,0,0,1024,1024,32,64,256]
hs_grid = [0,5,10,20,15,3,0,0]
```

```python
for i in range(8):
    bs = bs_grid[i]
    hs = hs_grid[i]
    meth = methods[i]    
    plt.figure(1, figsize = (12,12))
    if i ==0:
        lr = 0.01
        plt.plot(testacc[i],label='{}, lr = {}, bs = {}'.format(meth,lr,bs))
        
    elif bs != 0 and hs !=0:
        plt.plot(testacc[i],label='{}, bs = {}, hs = {}'.format(meth,bs,hs))
    elif bs != 0:
        plt.plot(testacc[i],label='{}, bs = {}'.format(meth,bs))
    elif hs != 0:
        plt.plot(testacc[i],label='{}, hs = {}'.format(meth,hs))
    else:
        plt.plot(testacc[i],label='{}'.format(meth))
plt.xlabel('Epochs')
plt.ylabel('Test Accuracy')
plt.title('Test accuracy in FCNN for different methods',fontsize = 20)
plt.legend(fontsize = 13)
plt.savefig("test_accuracy_fcnn.jpg")
plt.figure (2)
```
