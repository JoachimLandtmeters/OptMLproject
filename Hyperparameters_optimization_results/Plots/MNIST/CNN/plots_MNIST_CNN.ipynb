```python
import numpy as np
import matplotlib.pyplot as plt

```

# Test Accuracy


```python
testacc = np.loadtxt("Test_acccuracy.txt")

```

```python
np.shape(testacc)
```

```python
methods = ["SGD","LBFGSfullbatch","LBFGSNewminibatch","SDLBFGS", "Curveball"]

bs_grid = [64,0,1024,64,256]
hs_grid = [0,10,5,0,0]
epochs = [27, 15, 10, 30, 15]
```

```python
testacc[0,0:15]
```

```python
for i in range(5):
    bs = bs_grid[i]
    hs = hs_grid[i]
    meth = methods[i]    
    plt.figure(1, figsize = (12,12))
    if i ==0:
        lr = 0.01
        plt.plot(testacc[i,0:10],label='{}, lr = {}, bs = {}'.format(meth,lr,bs))
        
    elif bs != 0 and hs !=0:
        plt.plot(testacc[i,0:10],label='{}, bs = {}, hs = {}'.format(meth,bs,hs))
    elif bs != 0:
        plt.plot(testacc[i,0:10],label='{}, bs = {}'.format(meth,bs))
    elif hs != 0:
        plt.plot(testacc[i,0:10],label='{}, hs = {}'.format(meth,hs))
    else:
        plt.plot(testacc[i,0:10],label='{}'.format(meth))
plt.xlabel('Epochs')
plt.ylabel('Test Accuracy')
plt.title('Test accuracy in CNN for different methods',fontsize = 20)
plt.legend(fontsize = 13)
plt.savefig("test_accuracy_semilogy_cnn.jpg")
plt.figure (2)
```
