```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
testacc = np.loadtxt("LBFGS_minibatch_fcNN_testacc.txt")
trainacc = np.loadtxt("LBFGS_minibatch_fcNN_trainacc.txt")
testloss = np.loadtxt("LBFGS_minibatch_fcNN_testloss.txt")
trainloss = np.loadtxt("LBFGS_minibatch_fcNN_trainloss.txt")
```

```python
np.shape(testacc)
```

```python
for i in range(3):
    if  i==2:
        bs = 64
        hs = 10
    else:
        bs = 64
        hs = 20
        
    plt.figure(1, figsize = (12,10))
    plt.plot(testloss[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(2, figsize = (12,10))
    plt.plot(trainloss[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(3, figsize = (12,10))
    plt.plot(testacc[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(4, figsize = (12,10))
    plt.plot(trainacc[i],label='bs = {}, hs = {}'.format(bs,hs))
    plt.figure(1)
plt.xlabel('Epochs')
plt.ylabel('Test Loss')
plt.title('Minibatch LBFGS Test loss in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/mini_LBFGS_test_loss_fcnn.jpg")
plt.figure (2)
plt.xlabel('Epochs')
plt.ylabel('Train Loss')
plt.title('Minibatch LBFGS Train loss in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/mini_LBFGS_train_loss_fcnn.jpg")
plt.figure (3)
plt.xlabel('Epochs')
plt.ylabel('Test Accuracy')
plt.title('Minibatch LBFGS Test accuracy in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/mini_LBFGS_test_acc_fcnn.jpg")
plt.figure (4)
plt.xlabel('Epochs')
plt.ylabel('Train Accuracy')
plt.title('Minibatch LBFGS Train accuracy in FCNN for different batch sizes and history sizes',fontsize = 20)
plt.legend()
plt.savefig("best_results/mini_LBFGS_train_acc_fcnn.jpg")
```
