```python
# Building a sequential model 
# note : the order of the object in sequential is important
Items = []
linear = Linear(2, 25)
Items.append(linear)
Items.append(ReLu())
for i in range(hidden_layers-1):
    Items.append(Linear(25, 25))
    Items.append(ReLu())
Items.append(tanh())
Items.append(Linear(25, 2))
model = Sequential(Items)
```

```python
criterion = MSE()
optimizer = SGD(0.05)

epochs = 250
print_step = 25
mini_batch_size = 100

results["Custom Model"] = []
results["Siamese Model contrastive"].append(evaluate_model(SiameseCNN_Contrastive_Loss, nb_hidden_class,nb_hidden_comp, nb_samples, epochs, mini_batch_size, eta, L2_reg, validation_ratio, validation_rounds, True))

model_class, hidden_layer, nb_samples, criterion, optimizer, epochs, mini_batch_size, validation_ratio, validation_rounds, hide_output
train_model(model, train_input, train_target, criterion, optimizer, epochs, mini_batch_size, print_step)
```
